```python
import numpy as np   #Feature engineering for tree based Models
import pandas as pd
from sklearn.model_selection import KFold

def target_fold_encoding(train, test, cat_cols, target_col, n_splits=8, seed=42):
    
    
    
    train_encoded = train.copy()
    test_encoded = test.copy()

    global_means = train.groupby(cat_cols)[target_col].mean() if len(cat_cols)==1 else None

    for col in cat_cols:
        train_encoded[col + "_tfe"] = 0
        test_encoded[col + "_tfe"] = 0
        
        kf = KFold(n_splits=n_splits, shuffle=True, random_state=seed)

        # Empty array for OOF
        oof_vals = np.zeros(len(train))

        for train_idx, val_idx in kf.split(train):
            fold_tr, fold_val = train.iloc[train_idx], train.iloc[val_idx]

            # mean per category in fold
            fold_mean = fold_tr.groupby(col)[target_col].mean()

            # assign
            oof_vals[val_idx] = fold_val[col].map(fold_mean)

        # fill unseen categories with global mean
        global_mean = train[target_col].mean()
        oof_vals = np.where(np.isnan(oof_vals), global_mean, oof_vals)

        train_encoded[col + "_tfe"] = oof_vals

        # Test encodings using full-data mean
        full_mean = train.groupby(col)[target_col].mean()
        test_vals = test[col].map(full_mean)
        test_vals = test_vals.fillna(global_mean)

        test_encoded[col + "_tfe"] = test_vals

    return train_encoded, test_encoded


def create_interaction_features(df):
    
    """Creates high-value interaction features only (avoids noise):
    - Based on MI ranking you provided.
    - Avoids excessive polynomial blow-up."""
    

    new_df = df.copy()

    # Strong behavioral / lifestyle interactions
    new_df["activity_bmi_interaction"] = (
        new_df["physical_activity_minutes_per_week"] * (1 / (new_df["bmi"] + 1e-5))
    )

    new_df["waist_age"] = new_df["waist_to_hip_ratio"] * new_df["age"]
    new_df["bmi_age"] = new_df["bmi"] * new_df["age"]

    # Lipid interactions
    new_df["chol_hdl_ratio"] = (
        new_df["cholesterol_total"] / (new_df["hdl_cholesterol"] + 1e-5)
    )
    new_df["ldl_hdl_ratio"] = (
        new_df["ldl_cholesterol"] / (new_df["hdl_cholesterol"] + 1e-5)
    )
    new_df["tri_hdl_ratio"] = (
        new_df["triglycerides"] / (new_df["hdl_cholesterol"] + 1e-5)
    )

    # Blood pressure interactions
    new_df["pulse_pressure"] = new_df["systolic_bp"] - new_df["diastolic_bp"]
    new_df["bp_stress"] = new_df["systolic_bp"] * new_df["heart_rate"]

    # Lifestyle × health risk
    new_df["screen_sleep_ratio"] = (
        new_df["screen_time_hours_per_day"] / (new_df["sleep_hours_per_day"] + 1e-5)
    )

    new_df["alcohol_activity_interaction"] = (
        new_df["alcohol_consumption_per_week"]
        * new_df["physical_activity_minutes_per_week"]
    )

    # Add squared versions of top MI numeric features
    for feat in ["age", "waist_to_hip_ratio", "triglycerides"]:
        new_df[feat + "_sq"] = new_df[feat] ** 2

    return new_df


# -------------------------
# 🚀 USE THE FEATURE ENGINEERING HERE
# -------------------------

# Given:
# numeric_cols (20)
# categorical_cols (6)
# target = "diagnosed_diabetes"
# train, test must exist beforehand

categorical_cols = ['gender', 'ethnicity', 'education_level', 'income_level', 
                    'smoking_status', 'employment_status']

target_col = "diagnosed_diabetes"

# 1️⃣ Drop ID before feature engineering except keeping for merge/end-of-process
train_fe = train.copy()
test_fe = test.copy()

# 2️⃣ Target-Fold Encoding for categorical columns
train_fe, test_fe = target_fold_encoding(
    train_fe,
    test_fe,
    categorical_cols,
    target_col=target_col,
    n_splits=8,
    seed=42
)

# 3️⃣ Drop raw categorical columns, keep only encoded versions
train_fe = train_fe.drop(columns=categorical_cols)
test_fe = test_fe.drop(columns=categorical_cols)

# 4️⃣ Add meaningful interaction features
train_fe = create_interaction_features(train_fe)
test_fe = create_interaction_features(test_fe)

# 5️⃣ Remove leakage: remove target from test & interactions don't include target
if target_col in test_fe:
    test_fe = test_fe.drop(columns=[target_col])

# 6️⃣ Optionally drop ID before training (your main pipeline can handle this)
# train_fe = train_fe.drop(columns=["id"])
# test_fe = test_fe.drop(columns=["id"])

# train_fe, test_fe are now fully feature-engineered and ready for modeling
```

```python
# ============================================================
# 0. Imports
# ============================================================
import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.metrics import roc_auc_score

import xgboost as xgb
import optuna


# ============================================================
# 1. Target-Fold Encoding
# ============================================================
def target_fold_encoding(train, test, cat_cols, target_col, n_splits=8, seed=42):

    train_encoded = train.copy()
    test_encoded = test.copy()

    for col in cat_cols:
        train_encoded[col + "_tfe"] = 0.0
        test_encoded[col + "_tfe"] = 0.0

        kf = KFold(n_splits=n_splits, shuffle=True, random_state=seed)
        oof_vals = np.zeros(len(train))

        for tr_idx, val_idx in kf.split(train):
            fold_tr = train.iloc[tr_idx]
            fold_val = train.iloc[val_idx]

            fold_mean = fold_tr.groupby(col)[target_col].mean()
            oof_vals[val_idx] = fold_val[col].map(fold_mean)

        global_mean = train[target_col].mean()
        oof_vals = np.where(np.isnan(oof_vals), global_mean, oof_vals)

        train_encoded[col + "_tfe"] = oof_vals

        full_mean = train.groupby(col)[target_col].mean()
        test_vals = test[col].map(full_mean).fillna(global_mean)

        test_encoded[col + "_tfe"] = test_vals

    return train_encoded, test_encoded


# ============================================================
# 2. Interaction Feature Engineering
# ============================================================
def create_interaction_features(df):

    new_df = df.copy()

    new_df["activity_bmi_interaction"] = (
        new_df["physical_activity_minutes_per_week"] /
        (new_df["bmi"] + 1e-5)
    )

    new_df["waist_age"] = new_df["waist_to_hip_ratio"] * new_df["age"]
    new_df["bmi_age"] = new_df["bmi"] * new_df["age"]

    new_df["chol_hdl_ratio"] = (
        new_df["cholesterol_total"] /
        (new_df["hdl_cholesterol"] + 1e-5)
    )
    new_df["ldl_hdl_ratio"] = (
        new_df["ldl_cholesterol"] /
        (new_df["hdl_cholesterol"] + 1e-5)
    )
    new_df["tri_hdl_ratio"] = (
        new_df["triglycerides"] /
        (new_df["hdl_cholesterol"] + 1e-5)
    )

    new_df["pulse_pressure"] = (
        new_df["systolic_bp"] - new_df["diastolic_bp"]
    )
    new_df["bp_stress"] = (
        new_df["systolic_bp"] * new_df["heart_rate"]
    )

    new_df["screen_sleep_ratio"] = (
        new_df["screen_time_hours_per_day"] /
        (new_df["sleep_hours_per_day"] + 1e-5)
    )

    new_df["alcohol_activity_interaction"] = (
        new_df["alcohol_consumption_per_week"] *
        new_df["physical_activity_minutes_per_week"]
    )

    for feat in ["age", "waist_to_hip_ratio", "triglycerides"]:
        new_df[feat + "_sq"] = new_df[feat] ** 2

    return new_df


# ============================================================
# 3. Feature Engineering Pipeline
# ============================================================
categorical_cols = [
    'gender', 'ethnicity', 'education_level',
    'income_level', 'smoking_status', 'employment_status'
]

target_col = "diagnosed_diabetes"

train_fe = train.copy()
test_fe = test.copy()

train_fe, test_fe = target_fold_encoding(
    train_fe, test_fe, categorical_cols, target_col, n_splits=8, seed=42
)

train_fe = train_fe.drop(columns=categorical_cols)
test_fe = test_fe.drop(columns=categorical_cols)

train_fe = create_interaction_features(train_fe)
test_fe = create_interaction_features(test_fe)

if target_col in test_fe.columns:
    test_fe = test_fe.drop(columns=[target_col])


# ============================================================
# 4. Prepare Data
# ============================================================
X = train_fe.drop(columns=[target_col])
y = train_fe[target_col].values
X_test = test_fe.copy()


# ============================================================
# 5. Optuna Objective (OOF ROC-AUC)
# ============================================================
def objective(trial):

    params = {
        "objective": "binary:logistic",
        "eval_metric": "auc",
        "tree_method": "hist",
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.2, log=True),
        "max_depth": trial.suggest_int("max_depth", 3, 10),
        "min_child_weight": trial.suggest_int("min_child_weight", 1, 10),
        "subsample": trial.suggest_float("subsample", 0.6, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.6, 1.0),
        "gamma": trial.suggest_float("gamma", 0.0, 5.0),
        "reg_alpha": trial.suggest_float("reg_alpha", 0.0, 5.0),
        "reg_lambda": trial.suggest_float("reg_lambda", 0.0, 5.0),
        "n_estimators": trial.suggest_int("n_estimators", 300, 1200),
        "random_state": 42,
        "verbosity": 0
    }

    skf = StratifiedKFold(n_splits=8, shuffle=True, random_state=42)
    oof_preds = np.zeros(len(X))

    for tr_idx, val_idx in skf.split(X, y):
        X_tr, X_val = X.iloc[tr_idx], X.iloc[val_idx]
        y_tr, y_val = y[tr_idx], y[val_idx]

        model = xgb.XGBClassifier(**params)
        model.fit(X_tr, y_tr)

        oof_preds[val_idx] = model.predict_proba(X_val)[:, 1]

    return roc_auc_score(y, oof_preds)


# ============================================================
# 6. Run Optuna
# ============================================================
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50)

best_params = study.best_params
best_params.update({
    "objective": "binary:logistic",
    "eval_metric": "auc",
    "tree_method": "hist",
    "random_state": 42,
    "verbosity": 0
})


# ============================================================
# 7. Train Final Model + OOF + Test Predictions
# ============================================================
skf = StratifiedKFold(n_splits=8, shuffle=True, random_state=42)

oof_preds = np.zeros(len(X))
test_preds = np.zeros(len(X_test))

for tr_idx, val_idx in skf.split(X, y):
    X_tr, X_val = X.iloc[tr_idx], X.iloc[val_idx]
    y_tr, y_val = y[tr_idx], y[val_idx]

    model = xgb.XGBClassifier(**best_params)
    model.fit(X_tr, y_tr)

    oof_preds[val_idx] = model.predict_proba(X_val)[:, 1]
    test_preds += model.predict_proba(X_test)[:, 1] / 8


# ============================================================
# 8. Evaluation
# ============================================================
final_auc = roc_auc_score(y, oof_preds)
print(f"🔥 Final OOF ROC-AUC: {final_auc:.5f}")


# ============================================================
# 9. Submission File
# ============================================================
submission = pd.DataFrame({
    "id": test["id"],  # change if ID column name differs
    "diagnosed_diabetes": test_preds
})

submission.to_csv("submission.csv", index=False)
print("✅ submission.csv generated successfully!")  #You still hold promise, XGBoost
```
